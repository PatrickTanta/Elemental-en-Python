```python
from borracho import BorrachoTradicional
from campo import Campo
from coordenada import Coordenada


def caminata(campo, borracho, pasos):
    #primero definimos donde empieza
    inicio = campo.obtener_coordenada(borracho)
    origen = Coordenada(0,0)
    xs=[]
    ys=[]
    for _ in range(pasos):
        campo.mover_borracho(borracho)
        new_coord = campo.obtener_coordenada(borracho)
        xs.append(new_coord.x)
        ys.append(new_coord.y)
    return xs, ys


def simular_caminata(pasos, numero_de_intentos, tipo_de_borracho):
    borracho = tipo_de_borracho(nombre='Patrick')
    origen = Coordenada(0,0)

    for _ in range(numero_de_intentos):
        campo = Campo()
        campo.aadir_borracho(borracho, origen)  # pones al borracho en el origen
        coordsx, coordsy = caminata(campo, borracho, pasos)

    return coordsx, coordsy

def main(distancias_de_caminata, numero_de_intentos, tipo_de_borracho):
    distancias_para_plot = []
    for pasos in distancias_de_caminata:
        distancias_x, distancias_y = simular_caminata(pasos,numero_de_intentos, tipo_de_borracho)
    
    graficador(distancias_x, distancias_y)

# funcion para graficar
from bokeh.plotting import figure, show, output_notebook
output_notebook()


def graficador(x, y):
    grafica =figure(title='Camino aleatorio', x_axis_label='Coordenada X', y_axis_label='Coordenada Y', plot_width=800, plot_height=800)
    grafica.line(x, y, legend_label='recorrido')
    grafica.circle(x=x[0], y=y[0], color='blue', legend='inicio', size=10)
    grafica.circle(x=x[len(x)-1], y=y[len(x)-1], color='red', legend='fin', size=10)
    show(grafica)




distancias_de_caminata = [100000]
numero_de_intentos = 1

main(distancias_de_caminata, numero_de_intentos, BorrachoTradicional)
```
